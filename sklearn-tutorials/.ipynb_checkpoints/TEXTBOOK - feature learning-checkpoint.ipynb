```python
from sklearn import metrics
from sklearn.ensemble import ExtraTreesClassifier
import scipy as sp
import pandas as pd
import pylab as pl
from pandas.tools.plotting import scatter_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, f_classif

%pylab inline --no-import
```

```python

## where compactness = 4*pi*area / perimeter**2
features = ['area', 'perimeter', 'compactness', 'kernel_length', 
            'kernel_width', 'asymmetry_coeff', 'kernel_groove_len']
data = pd.read_csv('data/seeds.tsv', delimiter='\t', names=features+['types'])
print data.shape
```

```python
for feature in features:
    data.boxplot(column= feature, by = 'types', )
```

## It seems that most use features are 
- kernel_groove_len + kernel_length
- kernel_length
- perimeter
- area - very related to perimeter

```python
## tree model
X = np.array(data.loc[:, features])
le = LabelEncoder()
y = le.fit_transform(np.array(data.types))
print X.shape, y.shape
trees = ExtraTreesClassifier(n_estimators=200, max_features=5, bootstrap=True, random_state=0)
trees.fit(X, y)
feature_importances = pd.Series(trees.feature_importances_, index = features)
feature_importances.plot(kind = 'bar')
```

```python
selector = SelectKBest(k = 6, score_func=f_classif)
selector.fit(X, y)
feature_importances = pd.Series(selector.scores_, index = features)
feature_importances.plot(kind = 'bar')
pl.figure()
feature_pvalues = pd.Series(selector.pvalues_, index = features)
feature_pvalues.plot(kind = 'bar')
```
