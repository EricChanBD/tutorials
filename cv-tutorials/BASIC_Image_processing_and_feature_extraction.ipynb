***Motivation***

In state-of-the-art vision systems, the image pixel data are almost always *preprocessed* to form the measurement vector.

The philosophy behind image preprocessing is easy to understand; the image data may be contingent on many aspects of the real world that do not pertain to the task at hand.
For example, in an object detection task, the RGB values will change depending on the camera gain, illumination, object pose and particular instance of the object. The goal of image preprocessing is to **remove as much of this unwanted variation as possible**, while retaining the aspects of the image that are critical to the final decision.

**As such, it is usually important to take note of what kind of transformation a preprocessing step is invariant of**. 

It should be emphasized that preprocessing is very important; in practice, the choice of preprocessing technique can influence the performance of vision systems at least as much as the choice of model.

***Common Types of Image Preprocessing***

- **Per-pixel transfomration**, in which we replace each pixel in the image with a new value. These methods return a single value corresponding to each pixel of input image.
    - *Whitening* - the goal of whitening is to provide invariance to fluctuations in the mean intensity level and ctonrast of the image. Such variantion may arise because of a change in ambient lighting intensity, the object reflectance, or the camera gain. The mean and var are the global mean and var wrt the whole image.
For color images, this operation may be carried out by computing the mean and var from all three channels or by separately transforming each of channels based on their own statistics.
    - *Histogram equalization* - It is a nonlinear transformation that forces the distribution of pixel intensities to be flat. As a result, all the statistical moments of intensity values take predefined values. After this transformation, the intensities are equally distributed. E.g., in an image where many of the pixels are bright. Histogram equalization spreads these bright values out over a larger intensity range, and so has the effect of increasing the contrast in the brighter regions. The result is a continuous number rather than a discretized pixel intensity, but is in the same range as the original data. The result can be rounded to the nearest integer if subsequent processing demands. 
    - *Linear filtering* - After filtering an image, the new pixel value consists of a weighted sum of the intensities of pixels in the surrounding area of the original image. The weights are stored in a `filter kernel`. `zero padding` is usually used for border issues when convolve kernel with the image
        - `Gaussian (blurring) filter` - To blur an image, we convolve it with a 2D Gaussian, the blurring degree is dependent on the standard deviation of the Gaussian filter.
        - `First derivative filters and edge filters` - this type of filters locate places in the image where the intensity changes abruptly. E.g., `Prewitt and Sobel operators` can be used to detect horizontal and vertical edges
        - `Laplacian filters` - the discrete two dimensional approximation to the Laplacian operator (2nd derivative). Applying Laplacian filter to an image results in a response of high magnitude where the image is chaning, regardless of the direction of that change. The response is zero in regions that are flat and significant where edges occur in the image. It is hence invariant to constant additive changes in luminance, and useful for identifying interesting regions of the image.
        - `Laplacian of Gaussian filters` - In practice, the Laplacian operator produces noisy results. A superior approach is to first smooth the image with a Gaussian filter and then apply the Laplacian. Due to the associative property of convolution, we can equvalently convolve the Laplacian filter by a Gaussian and apply the resulting Laplacian of Gaussian filter to the image. It is parameterized by gaussian sigma, which can be tuned to be selective for changes at different scales. 
        - `Difference of Gaussians` - The Laplacian of Gaussian filter is very well approximated by the difference of Gaussians filter (compare figures 13.4d and 13.4e). As the name implies, this filter is created by taking the difference of two Gaussians at nearby scales. The same result can be achieved by filtering the image with the two Gaussians separately and taking the difference between the results.
        - `Gabor filters` - Gabor filters are selective for both scale and orientation. The 2D Gabor function is the product of a 2D Gaussian with a 2D sinusoid. It is parameterized by the Gaussian sigma, Sinusoid phase phi, orientation w, and wavelength lambda. *A set of Gabor filters with different frequencies and orientations may be helpful for extracting useful features from an image.*
        - `Harr-like filters`
- *Local binary patterns* - LBP returns a discrete value at each pixel that characterizeds the local texture in a way that is partially invariant to luminance. For this reason, features based on local binary patterns are commonly used as a substrate for face recognition algorithms. The LBP operators can be extended to use neighborhoods of different sizes: the central pixel is compared to positions in a circular pattern. In general, these positions must be estimated using bilinear interpolation. This extended LBP operator can capture texture at different scales in the image. Different methods of LBP exist in the literature, which can be gray-level invariant, rotation invariant or both.
    - *Texton maps*
- **Edges, corners and interest points**, in which we find and characeterize different types of interest points for local patterns/textures
    - *Candy edge detector*
    - *Harris corner detector*
    - *SIFT detector*
- **Descriptors**. Visual descriptors are low dimensional vectors that attempt to characterize the interesting aspects of an image *region* in a compact way
    - *Histograms*
    - *SIFT descriptors*
    - *HOG (histogram of oriented gradients) descriptors*
    - *Bag of words descriptor*
    - *Shape context descriptor*

```python
## Examples of whitening and histogram equalization
from skimage import io, color, exposure
import numpy as np
import pylab

def whiten(img):
    mu = np.mean(img)
    std = np.std(img)
    return (img - mu) / std

hawkes = io.imread('data/python-cv/hawkes_unequalized')
print hawkes.shape
whitened = whiten(hawkes)
hist_equalized = exposure.equalize_hist(hawkes)

fig, axes = pylab.subplots(nrows = 3, ncols = 1, figsize = (4, 4 * 3))
pylab.gray()
axes[0].imshow(hawkes)
axes[0].set_title('unequalized hawkes')
axes[1].imshow(whitened)
axes[1].set_title('whitened hawkes')
axes[2].imshow(hist_equalized)
axes[2].set_title('hist-equalized hawkes')
```

```python
## Gabor filter banks for texture classification
## online example: http://scikit-image.org/docs/dev/auto_examples/plot_gabor.html#example-plot-gabor-py
## Examples of how to classify textures based on Gabor filter banks

## Frequency and orientation representations of the Gabor filter 
## are similar to those of the human visual system. i.e., theta = (0, 45), frequency = (0.05, 0.25)

## The images are filtered using the real parts of various different Gabor filter kernels. 
## The mean and variance of the filtered images are then used as features for classification, 
## which is based on the least squared error for simplicity.

import numpy as np
from scipy import ndimage as nd
from skimage import data
from skimage.util import img_as_float
from skimage.filter import gabor_kernel

## feature extraction and match methods
def compute_feats(image, kernels):
    feats = np.zeros((len(kernels), 2), dtype = np.double)
    for k, kernel in enumerate(kernels):
        filtered = nd.convolve(image, kernel, mode = 'wrap', )
        feats[k] = filtered.mean(), filtered.var()
    return feats

def match(feats, ref_feats):
    """
    feats.shape = n_kernels x 2
    ref_feats.shape = n_objects x n_kernels x 2
    """
    return np.argmin([np.mean((feats - ref_feat)**2) for ref_feat in ref_feats])

## prepare filter bank kernels
kernels = [np.real(gabor_kernel(frequency = frequency, theta=theta, sigma_x=sigma, sigma_y=sigma)) 
            for theta in np.pi/4 * np.arange(4) 
            for sigma in (1, 3) 
            for frequency in (0.05, 0.25)]

## load images
shrink = (slice(0, None, 3), slice(0, None, 3))
brick = img_as_float(data.load('brick.png'))[shrink]
grass = img_as_float(data.load('grass.png'))[shrink]
wall = img_as_float(data.load('rough-wall.png'))[shrink]
image_names = ['brick', 'grass', 'wall']
images = [brick, grass, wall]

## prepare reference features
ref_feats = np.zeros((3, len(kernels), 2), dtype = np.double)
ref_feats[0, :, :] = compute_feats(brick, kernels)
ref_feats[1, :, :] = compute_feats(grass, kernels)
ref_feats[2, :, :] = compute_feats(wall, kernels)

#print ref_feats

print 'Rotated images matched against references using Gabor filter banks:'

print 'Original: brick, rotated: 180deg, match result:',
feats = compute_feats(nd.rotate(brick, angle = 180, reshape=False), kernels)
print image_names[match(feats, ref_feats)]

print 'Original: brick, rotated: 70deg, match result:',
feats = compute_feats(nd.rotate(brick, angle = 70, reshape=False), kernels)
print image_names[match(feats, ref_feats)]

print 'Original: grass, rotated: 145deg, match result:',
feats = compute_feats(nd.rotate(grass, angle = 145, reshape=False), kernels)
print image_names[match(feats, ref_feats)]

def power(image, kernel):
    ## normalize images for better comparison - whitening
    image = (image - image.mean()) / image.std()
    return np.sqrt(nd.convolve(image, np.real(kernel), mode = 'wrap') ** 2 + 
                   nd.convolve(image, np.imag(kernel), mode = 'wrap') ** 2)
## plot a selection of the filter bank kernels and their responses
results, kernel_params = [], []
for theta in (0, 1):
    theta = theta / 4. * np.pi
    for frequency in (0.1, 0.4):
        kernel = gabor_kernel(frequency, theta=theta)
        params = 'theta=%d,\nfrequency=%.2f' % (theta * 180 / np.pi, frequency )
        kernel_params.append(params)
        results.append((kernel, [power(img, kernel) for img in images]))
        
fig, axes = subplots(nrows = 6, ncols = 4, figsize = (9, 6))
gray()
fig.suptitle('Image responses for Gabor filter kernels', fontsize = 15)
axes[0][0].axis('off')

## plot original images
for label, img, ax in zip(image_names, images, axes[0][1:]):
    ax.imshow(img)
    ax.set_title(label)
    ax.axis('off')

for label, (kernel, powers), ax_row in zip(kernel_params, results, axes[1:-1]):
    ## plot Gabor kernels
    ax = ax_row[0]
    ax.imshow(np.real(kernel), interpolation='nearest')
    ax.set_ylabel(label, fontsize=7)
    ## plot Gabor responses with the contrast normalized for each filter
    vmin = np.min(powers)
    vmax = np.max(powers)
    for patch, ax in zip(powers, ax_row[1:]):
        ax.imshow(patch, )#vmin=vmin, vmax=vmax, )
        ax.axis('off')
## reconstructed images
axes[-1, 0].axis('off')
for ax, (kernel, powers) in zip(axes[-1][1:], results):
    recontructed = (powers[0]+powers[1]+powers[2]) / len(powers)
    print 
    ax.imshow(recontructed)
```

```python
## Gabor filter
## frequency : float, Frequency of the harmonic function.
## theta : float, Orientation in radians. If 0, the harmonic is in the x-direction.
## bandwidth : float, The bandwidth captured by the filter. For fixed bandwidth, `sigma_x`
##      and `sigma_y` will decrease with increasing frequency. This value is
##      ignored if `sigma_x` and `sigma_y` are set by the user.
## sigma_x, sigma_y : float, Standard deviation in x- and y-directions. These directions apply to
##      the kernel *before* rotation. If `theta = pi/2`, then the kernel is
##      rotated 90 degrees so that `sigma_x` controls the *vertical* direction.
## offset : float, optional, Phase offset of harmonic function in radians.
from math import pi
rice = 255-color.rgb2gray(io.imread('data/python-cv/rice.jpg'))
fig, axes = subplots(nrows = 3, ncols = 3, figsize = (12, 12))
axes[0, 0].imshow(rice)
axes[0, 0].set_title('rice')
#thetas = [0, 45, 90, 135, 180, 225, 270, 270+45]
reconstructed = np.zeros_like(rice)
thetas = [0, pi/4, pi/2, pi*3/4, pi, pi*5/4, pi*6/4, pi*7/4,]
for i in [0, 1, 2]:
    for j in [0, 1, 2]:
        if i * 3 + j >= len(thetas):
            break
        theta = thetas[i*3+j]
        gabors_0_real, gabors_0_img = filter.gabor_filter(whiten(rice), frequency = .05, 
            theta=theta, sigma_x= None, sigma_y = None)
        gabors_img = np.sqrt(gabors_0_real**2 + gabors_0_img**2)
        reconstructed = reconstructed + gabors_img
        axes[i, j].imshow(gabors_img)
        axes[i, j].set_title(r'gabor with $freq=.05\theta=%f$' % theta)
axes[-1, -1].imshow(rice)
axes[-1, -1].set_title('original')

figure()
imshow(reconstructed / len(thetas))
```

```python
## Texture Classification based on Local Binary Pattern
## using skimage library
## http://scikit-image.org/docs/dev/auto_examples/plot_local_binary_pattern.html
## In this example, we will see how to classify textures based on LBP (Local Binary Pattern). 
## The histogram of the LBP result is a good measure to classify textures. 
## For simplicity the histogram distributions are then tested against each other 
## using the Kullback-Leibler-Divergence.

from skimage import transform
from skimage import feature
from skimage import data
from skimage import util

## Load images
brick = data.load('brick.png')
grass = data.load('grass.png')
wall = data.load('rough-wall.png')
images = [brick, grass, wall]
image_names = ['brick', 'grass', 'wall']

## Roated images
rotated_brick = transform.rotate(brick, angle = 60, resize = False)
rotated_grass = transform.rotate(grass, angle = 75, resize = False)
rotated_wall = transform.rotate(wall, angle = 125, resize = False)
rotates = [rotated_brick, rotated_grass, rotated_wall]

## settings for LBP
METHOD = 'uniform' ## rotation invariant features!?
P, R = 16, 2
## LBP for original images
lbp_brick = feature.local_binary_pattern(brick, P = P, R = R, method = METHOD)
lbp_grass = feature.local_binary_pattern(grass, P = P, R = R, method = METHOD)
lbp_wall = feature.local_binary_pattern(wall, P = P, R = R, method = METHOD)
lbps = [lbp_brick, lbp_grass, lbp_wall]

## LBP for rotated images
lbp_rotated_brick = feature.local_binary_pattern(rotated_brick, P = P, R = R, method = METHOD)
lbp_rotated_grass = feature.local_binary_pattern(rotated_grass, P = P, R = R, method = METHOD)
lbp_rotated_wall = feature.local_binary_pattern(rotated_wall, P = P, R = R, method = METHOD)
rotated_lbps = [lbp_rotated_brick, lbp_rotated_grass, lbp_rotated_wall]

## draw patterns vs original images
fig, axes = subplots(nrows = 3, ncols = 4, figsize = (4 * 4, 4 * 3))
fig.suptitle('Image v.s. Local Binary Pattern')
for (ax, img_name, img, lbp_img, rotated, lbp_rotated) in zip(axes, image_names, images, lbps, rotates, rotated_lbps):
    ax[0].imshow(img)
    ax[0].set_ylabel(img_name)
    ax[1].imshow(lbp_img)
    ax[2].imshow(rotated)
    ax[3].imshow(lbp_rotated)
    
## texture classification
def kl_divergence(img1, img2):
    #img1 = util.img_as_float(img1)
    #img2 = util.img_as_float(img2)
    p, _ = np.histogram(img1, density = True)
    q, _ = np.histogram(img2, density = True)
    common = np.logical_and(p != 0, q != 0)
    return np.sum(p[common] * np.log2(p[common]/q[common]))
    #return np.sum(p * np.log2(p/q))
    
## matching based on original images
"""
figure()
fig, axes = subplots(nrows = 2, ncols = 3, figsize = (4 * 3, 4 * 2))
fig.suptitle('histogram raw images')
for i in xrange(3):
    _ = axes[0, i].hist(images[i], normed = True, )
    axes[0, i].set_title(image_names[i])
    _ = axes[1, i].hist(rotates[i], normed = True, )
    
figure()
fig, axes = subplots(nrows = 2, ncols = 3, figsize = (4 * 3, 4 * 2))
fig.suptitle('histogram lbps')
for i in xrange(3):
    _ = axes[0, i].hist(lbps[i], normed = True, bins = 50)
    axes[0, i].set_title(image_names[i])
    _ = axes[1, i].hist(rotated_lbps[i], normed = True, bins = 50)

"""

for img_name1, img in zip(image_names, images):
    for img_name2, rotated in zip(image_names, rotates):
        print 'ORIGINAL: %s vs %s = %g' % (img_name1, img_name2, kl_divergence(util.img_as_int(img), util.img_as_int(rotated)))
    print ''
## matching based on lbp images
for img_name1, img in zip(image_names, lbps):
    for img_name2, rotated in zip(image_names, rotated_lbps):
        print 'LBP: %s vs %s = %g' % (img_name1, img_name2, kl_divergence(img, rotated))
    print ''
```
