```python
## load data
train_digits, valid_digits, test_digits = cPickle.load(open('data/mnist.pkl'))
train_X, train_y = train_digits
valid_X, valid_y = valid_digits
test_X, test_y = test_digits

X = np.vstack([train_X, valid_X, test_X])
y = np.concatenate([train_y, valid_y, test_y], )

print train_X.shape, valid_X.shape, test_X.shape
print y.shape
print X.min(), X.max()
```

```python
## pairwise PCA of each pair of classes
from itertools import product
fig, axes = pylab.subplots(nrows = 10, ncols = 10, figsize=(20, 20))
pylab.prism()


pca = RandomizedPCA(n_components=2)

for (i, j) in product(range(10), repeat = 2):
    if i > j:
        continue
    index = (y == i) + (y == j)
    X_ = X[index]
    y_ = y[index]
    X_transformed = pca.fit_transform(X_)
    axes[i, j].scatter(X_transformed[:,0], X_transformed[:, 1], c=y_, marker='+', s=50)
    axes[i, j].get_xaxis().set_visible(False)
    axes[i, j].get_yaxis().set_visible(False)
    axes[j, i].scatter(X_transformed[:,0], X_transformed[:, 1], c=y_, marker='+', s=50)
    axes[j, i].get_xaxis().set_visible(False)
    axes[j, i].get_yaxis().set_visible(False)
pylab.tight_layout()
```

```python
## Classification with Kernel approximation Nystroem and RBFSampler
## gamma measures the span of local neigbhor - which corresponds to about 6 pixels 
nystroem_feat_extractor = Nystroem(gamma = 0.012, n_components=400)
fourier_feat_extractor = RBFSampler(gamma = 0.012, n_components=400)
#sgd = SGDClassifier(loss = 'hinge', penalty = 'l2', alpha = 0.01) # C = 100

nystroem_sgd = Pipeline(steps = [('nystroem', nystroem_feat_extractor)
                                 , ('sgd', SGDClassifier(loss = 'log', 
                                                         penalty = 'l2', 
                                                         alpha = 1e-4))])
fourier_sgd = Pipeline(steps = [('fourier', fourier_feat_extractor)
                                 , ('sgd', SGDClassifier(loss = 'log', 
                                                         penalty = 'l2', 
                                                         alpha = 1e-4))])

%time nystroem_sgd.fit(train_X, train_y)
%time fourier_sgd.fit(train_X, train_y)

print 'nystroem performance'
print nystroem_sgd.score(train_X, train_y)
print nystroem_sgd.score(valid_X, valid_y)
print nystroem_sgd.score(test_X, test_y)

print 'fourier performance'
print fourier_sgd.score(train_X, train_y)
print fourier_sgd.score(valid_X, valid_y)
print fourier_sgd.score(test_X, test_y)
```

### how to do feature extraction Correctly in sklearn

Conclusion drawn from the source

- The baseline linear model (SVM with squared hinge loss and l2 regularization from liblinear) has ~91% test accuracy when trained on 20k samples.
- The 1000-dim k-means based feature expansion should yield ~96% test accuracy in less than 20s (unsupervised feature extraction + classifier training) on the same 20k samples.
- The 1000-dim random sample based feature expansion should yield ~96% test accuracy in less than 10s (unsupervised feature extraction + classifier training) on the same 20k samples.
- Deep architectures based on those feature extraction models do not seem the be able to improve upon those accurracy numbers on this dataset.

## Soft threshold k-means features
- it is a cheap kind of kernel expansion alternative, it is called poor man's yet scalable alternative to Sparse Coding based feature extraction
- two main implementations here: soft-threshold k-means & soft-threshold random-sampling

```python
## scalable version of machine learning tools
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import normalize
from sklearn.decomposition import RandomizedPCA
from sklearn.pipeline import Pipeline
from sklearn.utils import check_random_state
## Linear SVC scales with n_samples almost linearly
## another choice is to use SGDClassifier, which can even
## be trained out-of-core
from sklearn.svm import LinearSVC
```

```python
def soft_threshold(data, centers, threshold = 0.0):
    ## compute cosine similarites of samples wrt kmeans centers
    ## normalize along axis = 1, normalization by SAMPLE
    C = normalize(centers)
    X = normalize(data)
    similarities = np.dot(X, C.T)
    
    ## remove negative cosine features (~50% of them)
    similarities[similarities < threshold] = 0.0
    
    ## project the new features on the unit euclidean ball
    ## NOT sure if it is really reasonable to normalize
    #return normalize(similarities, copy=True)
    return similarities

class SamplingMapper(BaseEstimator, TransformerMixin):
    def __init__(self, n_components, threshold = 0.0, 
                 random_state = None):
        self.n_components = n_components
        self.threshold = threshold
        self.random_state = random_state
    def fit(self, X, y = None):
        """
        Randomly sample support vectors
        """
        random_state = check_random_state(self.random_state)
        X = np.asarray(X)
        sv_indices = np.arange(X.shape[0])
        random_state.shuffle(sv_indices)
        self.sv_indices_ = sv_indices[:self.n_components]
        self.components_ = X[self.sv_indices_]
        return self
    def transform(self, X):
        return soft_threshold(X, self.components_, 
                              threshold = self.threshold)
    
class KMeansMapper(BaseEstimator, TransformerMixin):
    def __init__(self, n_components, threshold = 0.0, 
                 *args, **kwargs):
        self.n_components = n_components
        self.threshold = threshold
        self.model_ = MiniBatchKMeans(n_clusters=self.n_components, 
                                      *args, **kwargs)
    def fit(self, X, y = None):
        self.model_.fit(X)
        return self
    def transform(self, X):
        return soft_threshold(X, self.model_.cluster_centers_, 
                              threshold = self.threshold)
```

```python
def make_sst_model(n_features, n_pca = 50, threshold = 0.0):
    """
    create a pipeline model of sampling soft-threshold
    feature learner
    First Use PCA for to reduce the dimensionality to make 
    distance calculation/clustering converge faster
    """
    return Pipeline(steps = [
        ('dim_reduction', RandomizedPCA(n_components=n_pca, whiten = True, ))
        , ('feature_map', SamplingMapper(n_components = n_features, threshold = threshold))
        #, ('clf', SGDClassifier(loss = 'log', penalty='elasticnet', alpha = 1e-4, n_jobs=-1))
        , ('clf', LinearSVC(C = 1))
    ])

n_features_range = [30, 100, 300, 1000, 3000]
train_scores = [None] * len(n_features_range)
valid_scores = [None] * len(n_features_range)
test_scores = [None] * len(n_features_range)

for i, n_features in enumerate(n_features_range):
    print 'testing', n_features
    sst = make_sst_model(n_features)
    sst.fit(train_X, train_y)
    train_scores[i] = sst.score(train_X, train_y)
    valid_scores[i] = sst.score(valid_X, valid_y)
    test_scores[i] = sst.score(test_X, test_y)
    print train_scores[i], valid_scores[i], test_scores[i]
```

```python
def make_kmst_model(n_features, n_pca = 50, threshold = 0.0):
    """
    create a pipeline model of sampling soft-threshold
    feature learner
    First Use PCA for to reduce the dimensionality to make 
    distance calculation/clustering converge faster
    """
    return Pipeline(steps = [
        ('dim_reduction', RandomizedPCA(n_components=n_pca, whiten = True, ))
        , ('feature_map', KMeansMapper(n_components = n_features, threshold = threshold, 
                                       batch_size = 1000, init='random'))
        #, ('clf', SGDClassifier(loss = 'log', penalty='elasticnet', alpha = 1e-4, n_jobs=-1))
        , ('clf', LinearSVC(C = 1))
    ])

n_features_range = [30, 100, 300, 500, 1000, 1500, 3000, 4000]
train_scores = [None] * len(n_features_range)
valid_scores = [None] * len(n_features_range)
test_scores = [None] * len(n_features_range)

for i, n_features in enumerate(n_features_range):
    print 'testing', n_features
    kmst = make_kmst_model(n_features)
    kmst.fit(train_X, train_y)
    train_scores[i] = kmst.score(train_X, train_y)
    valid_scores[i] = kmst.score(valid_X, valid_y)
    test_scores[i] = kmst.score(test_X, test_y)
    print train_scores[i], valid_scores[i], test_scores[i]
```

```python
## scalable version of machine learning tools
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import normalize
from sklearn.decomposition import RandomizedPCA
from sklearn.pipeline import Pipeline
from sklearn.utils import check_random_state
## Linear SVC scales with n_samples almost linearly
## another choice is to use SGDClassifier, which can even
## be trained out-of-core
from sklearn.svm import LinearSVC

def soft_threshold(data, centers, threshold = 0.0):
	## normalize both data and centers PER SAMPLE
	C = normalize(centers)
	X = normalize(data)
	## calculate cosine distances 
	similarities = np.dot(X, C.T)
	nsamples, ncenters = similarities.shape
	feats = np.empty((nsamples, 2*ncenters))

	## thresholding - pos and neg
	feats[:, :ncenters] = similarities
	feats[similarities<threshold, :ncenters] = 0.0
	feats[:, ncenters:] = -similarities
	feats[-similarities<threshold, ncenters:] = 0.0
	return feats

class SamplingMapper(BaseEstimator, TransformerMixin):
    def __init__(self, n_components, threshold = 0.0, 
                 random_state = None):
        self.n_components = n_components
        self.threshold = threshold
        self.random_state = random_state
    def fit(self, X, y = None):
        """
        Randomly sample support vectors
        """
        random_state = check_random_state(self.random_state)
        X = np.asarray(X)
        sv_indices = np.arange(X.shape[0])
        random_state.shuffle(sv_indices)
        self.sv_indices_ = sv_indices[:self.n_components]
        self.components_ = X[self.sv_indices_]
        return self
    def transform(self, X):
        return soft_threshold(X, self.components_, 
                              threshold = self.threshold)
    
class KMeansMapper(BaseEstimator, TransformerMixin):
    def __init__(self, n_components, threshold = 0.0, 
                 *args, **kwargs):
        self.n_components = n_components
        self.threshold = threshold
        self.model_ = MiniBatchKMeans(n_clusters=self.n_components, 
                                      *args, **kwargs)
    def fit(self, X, y = None):
        self.model_.fit(X)
        return self
    def transform(self, X):
        return soft_threshold(X, self.model_.cluster_centers_, 
                              threshold = self.threshold)
```
