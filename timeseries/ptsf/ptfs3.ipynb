```python
%pylab inline --no-import-all
import pylab as pl
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from datetime import datetime
from sklearn.metrics import r2_score
```

## Forecasting Method
- Model Based (*strong assumption, working better when data is sparse*):
    - multi-regression analysis
    - autoregression analysis
    - logistic regression 
- Data Driven:
    - smoothing method (SES, Holt-Winter)
    - NN
    - CART
- Combination:
    - ensemble
    - boosting (ensemble on error)
    
Model-Based: global trend

Data-Driven: local trend (short memory param) + global trend (long memory param)

###1. Regression Based Model
- linear (exp, poly) regression - **for trend**
- additive/multiplicative seasonal component - **for seasonal**
- autoregression for improvement
- autoregression for forecastability

#### regression for trend

```python
def as_yearmon(month_year):
    dt = datetime.strptime(month_year, "%b-%y")
    return pd.Period(year = dt.year, month = dt.month, freq='M')
```

```python
amtrack = pd.read_csv("../../data/time-series/Amtrak.csv", )
amtrack = amtrack.iloc[:, :2]
amtrack.Month = map(as_yearmon, amtrack.Month)
amtrack.set_index("Month", inplace=True)
```

```python
amtrack.plot()
```

```python
X = amtrack.copy()
X['t'] = np.arange(1, X.shape[0]+1)
#X = sm.add_constant(X)
print X.head()
test_X = X.iloc[-12:, :]
train_X = X.iloc[:-12, :]
print train_X.shape, test_X.shape
```

```python
## linear regression
model = smf.ols('Ridership ~ t', df = train_X)
model = model.fit()
print model.summary()

print r2_score(train_X.Ridership, model.predict(train_X ))
print r2_score(test_X.Ridership, model.predict(test_X ))

ax = X.Ridership.plot()
ax.plot_date(X.index, model.predict(X), 'r-')
```

```python
## exp regression
model = smf.ols('np.log(Ridership) ~ t', df = train_X)
model = model.fit()
print model.summary()

print r2_score(train_X.Ridership, np.exp(model.predict(train_X )))
print r2_score(test_X.Ridership, np.exp(model.predict(test_X )))

ax = X.Ridership.plot()
ax.plot_date(X.index, np.exp(model.predict(X)), 'r-')
```

```python
## poly regression
model = smf.ols('Ridership ~ t + I(t**2)', df = train_X)
model = model.fit()
print model.summary()

print r2_score(train_X.Ridership, model.predict(train_X ))
print r2_score(test_X.Ridership, model.predict(test_X ))

ax = X.Ridership.plot()
ax.plot_date(X.index, model.predict(X), 'r-')

pl.figure()
pl.plot(X.Ridership - model.predict(X), label = 'Residual')
pl.legend()

_ = sm.graphics.tsa.plot_acf(X.Ridership - model.predict(X), lags = 10, )
```

*2nd order Polynomial works quite well with trend, but the acf of residuals at lag 6 reveals there is a seasonal part*

*and acf at lag = 1 reveals an autogression model*

#### use dummy variable to model seasonality

```python
season = map(lambda t: t.month, X.index)
XX = pd.concat([X, pd.get_dummies(season, prefix="M").set_index(X.index)], axis = 1, )
print XX.shape
print XX.columns

train_X = XX.iloc[:-12, :]
test_X = XX.iloc[-12:, :]
print train_X.shape, test_X.shape
```

```python
## poly regression + season
model = smf.ols('Ridership ~ t + I(t**2) + M_1+M_2+M_3+M_4+M_5+M_6+M_7+M_8+M_9+M_10+M_11+M_12', df = train_X)
model = model.fit()
print model.summary()

print r2_score(train_X.Ridership, model.predict(train_X ))
print r2_score(test_X.Ridership, model.predict(test_X ))

ax = X.Ridership.plot()
ax.plot_date(XX.index, model.predict(XX), 'r-')

pl.figure()
pl.plot(XX.Ridership - model.predict(XX), label = 'Residual')
pl.legend()

_ = sm.graphics.tsa.plot_acf(XX.Ridership - model.predict(XX), lags = 15, )
```

*Now we only got autocorrelation left in the residual*

*And it doesnt see to be an multiplicative model*

```python
## poly regression + season - multiplicative model
model = smf.ols('np.log(Ridership) ~ t + I(t**2) + M_1+M_2+M_3+M_4+M_5+M_6+M_7+M_8+M_9+M_10+M_11+M_12', df = train_X)
model = model.fit()
print model.summary()

print r2_score(train_X.Ridership, np.exp(model.predict(train_X )))
print r2_score(test_X.Ridership, np.exp(model.predict(test_X )))

ax = X.Ridership.plot()
ax.plot_date(XX.index, np.exp(model.predict(XX)), 'r-')

pl.figure()
pl.plot(XX.Ridership - np.exp(model.predict(XX)), label = 'Residual')
pl.legend()

_ = sm.graphics.tsa.plot_acf(XX.Ridership - np.exp(model.predict(XX)), lags = 15, )
```

```python
## or use sin/cos to approximate continous seasonality function
model = smf.ols('Ridership ~ t + I(t**2) + np.sin(np.pi*2*t/52.18) + np.cos(np.pi*2*t/52.18)', df = train_X)
model = model.fit()
print model.summary()

print r2_score(train_X.Ridership, model.predict(train_X ))
print r2_score(test_X.Ridership, model.predict(test_X ))

ax = X.Ridership.plot()
ax.plot_date(X.index, model.predict(X), 'r-')

pl.figure()
pl.plot(X.Ridership - model.predict(X), label = 'Residual')
pl.legend()

_ = sm.graphics.tsa.plot_acf(X.Ridership - model.predict(X), lags = 15, )
```
