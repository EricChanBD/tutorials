## Forecasting Method
- Model Based (*strong assumption, working better when data is sparse*):
    - multi-regression analysis
    - autoregression analysis
    - logistic regression 
- Data Driven:
    - smoothing method (SES, Holt-Winter)
    - NN
    - CART
- Combination:
    - ensemble
    - boosting (ensemble on error)
    
Model-Based: global trend

Data-Driven: local trend (short memory param) + global trend (long memory param)

- linear (exp, poly) regression - **for trend**
- additive/multiplicative seasonal component - **for seasonal**

```python
def as_yearmon(month_year):
    dt = datetime.strptime(month_year, "%b-%y")
    return pd.Period(year = dt.year, month = dt.month, freq='M')
```
