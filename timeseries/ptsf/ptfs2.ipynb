### Sales Souvenir

```python
sales = pd.read_csv("../../data/time-series/SouvenirSales.csv")
def as_yearmon(month_year):
    dt = datetime.strptime(month_year, '%b-%y')
    return pd.Period(year = dt.year, month=dt.month, freq='M')
sales.Date = map(as_yearmon, sales.Date)
sales.set_index("Date", inplace=True)
```

```python
ax = sales.plot()
pd.rolling_mean(sales, 4, ).plot(ax = ax, )
```

```python
train_sales = sales.loc[:"Jan 2001", :]
test_sales = sales.loc["Jan 2001":, :]
print train_sales.shape, test_sales.shape
```

```python
## simple prediction 
yhat = test_sales.copy()
yhat.iloc[:, :] = train_sales.iloc[-1, 0]
print 'RMSE:', np.sqrt(mean_squared_error(test_sales, yhat))
print 'MAPE:', mape(test_sales, yhat)
```

```python
ax = test_sales.plot()
yhat.plot(ax=ax)
ax.plot_date(test_sales.index, yhat - test_sales, 'r', label="residual")
ax.legend()
```
