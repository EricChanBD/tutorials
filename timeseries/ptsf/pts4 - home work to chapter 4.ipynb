```python
%pylab inline --no-import-all
import pylab as pl
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from datetime import datetime
```

```python
def yearmon(month_year):
    dt = datetime.strptime(month_year, "%b-%y")
    return pd.Period(year = dt.year, month = dt.month, freq='M')
```

```python
sep11 = pd.read_csv("../../data/time-series/Sept11Travel.csv", thousands=",")
sep11.Month = map(yearmon, sep11.Month)
sep11.set_index("Month", inplace = True)
```

```python
_ = sep11.plot(subplots=True)
_ = sm.graphics.tsa.plot_acf(sep11['Air RPM'], lags = 20)
```
