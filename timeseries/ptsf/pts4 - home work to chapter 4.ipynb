### Toys R Revenue

```python
def quarter_year(qy):
    q, y = map(int, qy[1:].split("-"))
    return pd.Period(year = y, quarter = q, freq = 'Q')

toys = pd.read_csv("../../data/time-series/ToysRUsRevenues.csv")
toys.QuarterYear = map(quarter_year, toys.QuarterYear)
toys.set_index("QuarterYear", inplace = True)
```

```python
toys.plot()

_ = sm.graphics.tsa.plot_acf(toys.Revenue)
```

*trend (linear?) + multiplicative seasonal components, not really auto-regression*

```python
X = toys.copy()
X['t'] = np.arange(X.shape[0])
season = pd.get_dummies(map(lambda x: x.quarter, toys.index), prefix="Q").set_index(X.index)
X = pd.concat([X, season], axis=1)
print X.columns
```

```python
model = smf.ols("Revenue ~ t + Q_1+Q_2+Q_3+Q_4 ", df = X).fit()
print model.summary()
residual = X.Revenue - model.predict(X)
ax = X.Revenue.plot()
ax.plot_date(X.index, model.predict(X), 'r-', label = 'prediction')
ax.legend()
ax.set_title("linear + additive season")
_ = sm.graphics.tsa.plot_acf(residual)
```

```python
model = smf.ols("np.log(Revenue) ~ t + Q_1+Q_2+Q_3+Q_4 ", df = X).fit()
print model.summary()
residual = X.Revenue - np.exp(model.predict(X))
ax = X.Revenue.plot()
ax.plot_date(X.index, np.exp(model.predict(X)), 'r-', label = 'prediction')
ax.legend()
ax.set_title("linear + multiplicative season")
_ = sm.graphics.tsa.plot_acf(residual)
```

*multiplicative model works better than additive model*

### Walmart Stock

```python
def date(dt):
    dt = datetime.strptime(dt, "%d-%b-%y")
    return pd.Period(freq="D", year=dt.year, month=dt.month, day = dt.day)

walmart = pd.read_csv("../../data/time-series/WalMartStock.csv")
walmart.Date = map(date, walmart.Date)
walmart.set_index("Date", inplace = True)
```

```python
walmart.plot()
walmart.diff().plot()
_ = sm.graphics.tsa.plot_acf(walmart.Close, lags = 90)
_ = sm.graphics.tsa.plot_pacf(walmart.Close, lags = 90)
```

*strongly suspect it is a random walk?*

```python
X = walmart.copy()
X['lag1'] = X.shift()
```

```python
model = smf.ols("Close ~ lag1", df = X.iloc[1:, :]).fit()
print model.summary()
```

*not really random walk, as 95% of coef for lag1 does not cover 1, but it is very close*

### Department sales

```python
sales = pd.read_csv("../../data/time-series/DepartmentStoreSales.csv")
sales.set_index("Quarter", inplace = True)
```

```python
sales.plot()

_ = sm.graphics.tsa.plot_acf(sales.Sales)
```

*multiplicative seasonal (4 months) + linear trend*

```python
X = sales.copy()
X['t'] = np.arange(X.shape[0])
seasons = pd.get_dummies(map(lambda q: q%4, X.index), prefix = 'S').set_index(X.index)
X = pd.concat([X, seasons], axis = 1)
```

```python
model = smf.ols("np.log(Sales) ~ t + S_0+S_1+S_2+S_3", X).fit()
print model.summary()
residual = X.Sales - np.exp(model.predict(X))



ax = X.Sales.plot()
ax.plot(X.index, np.exp(model.predict(X)), label = 'prediction', color = 'r')
ax.legend()

_ = sm.graphics.tsa.plot_acf(residual)
```
