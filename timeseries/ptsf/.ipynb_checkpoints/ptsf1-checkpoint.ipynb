## quarterly sales

```python
sales = pd.read_csv("../../data/time-series/DepartmentStoreSales.csv")
sales = sales.set_index("Quarter")

cycles, trend = sm.tsa.filters.hpfilter(sales, lamb=1600, )
_ = pd.DataFrame(dict(sales = sales.Sales, trend = trend, cycles = cycles), index=sales.index).plot(subplots=True)
```

```python
rsales = com.convert_to_r_dataframe(sales)
%Rpush rsales
%R print(names(rsales))
%R rsales <- ts(rsales$Sales, start = 1, frequency=4)
%R plot(decompose(rsales, type = 'additive'))
%R plot(decompose(rsales, type = 'multiplicative'))
```

## quarterly shipment

```python
shipment = pd.read_csv("../../data/time-series/ApplianceShipments.csv")
shipment.Quarter = shipment.Quarter.apply(lambda s: pd.Period('Q'.join(s[1:].split("-"))))
shipment = shipment.set_index("Quarter")
shipment.sort_index(inplace=True)

cycles, trend = sm.tsa.filters.hpfilter(shipment, lamb=1600, )
_ = pd.DataFrame(dict(shipment = shipment.Shipments, 
                      trend = trend, 
                      cycles = cycles), 
                 index=shipment.index).plot(subplots=True, figsize=(16, 8))
```

```python
rshipment = com.convert_to_r_dataframe(shipment)
%Rpush rshipment
%R rshipment <- ts(rshipment$Shipments, start = c(1985, 1), frequency=4)
%R plot(decompose(rshipment))
%R result <- decompose(rshipment)
%Rget --as_dataframe result
```

```python
%%R
require(zoo)
rshampoon <- read.csv("../../data/temp/rshampoon.csv", stringsAsFactors=F)
rshampoon$Month = as.yearmon(rshampoon$Month)
rshampoon <- zoo(rshampoon$Shampoo.Sales, rshampoon$Month)

plot(rshampoon)
plot(decompose(as.ts(rshampoon, type="multiplicative")))
plot(stl(rshampoon, s.window='periodic'))
```
