# Feature Engineering

```python
## tri-kmeans feature selection
class TriKmeansFeatures(BaseEstimator):
    def __init__(self, n_clusters, n_batches, batch_size_range, cache_dir, random_state=0):
        """
        n_clusters: number of clusters used in KMeans
        n_batches: number of batches extracted from original feats
        batch_size_range: size range of batch including how many original feats [low,high]
        cache_dir: the cache directory for shared memory object
        So the total extract features would be n_batches * n_clusters.
        And the batches will be extracted randomly with replacement from origna
        feats
        Clustering methods are sensitive to feature scales. Its better to normalize 
        the train and test matrix X before feeding it to clustering methods, such as
        KMeans and MiniBatchKMeans
        """
        self.n_clusters = n_clusters
        self.n_batches = n_batches
        self.batch_size_range = batch_size_range
        self.cache_dir = cache_dir
        #self.rng = check_random_state(random_state)
        #np.random.set_state(self.rng)
        random.seed(random_state)
    def fit(self, X, y=None):
        """FITTING STEPS:
        1. extract batches from original features (without redundancy)
        2. for each batch, train a clustering model, in parallel
        """
        ## 1. extract random features
        ## chance is thr that same batch of features are selected, 
        ## but it is small when n_features is large
        n_samples, n_features = X.shape
        batch_indices = []
        while len(batch_indices) != self.n_batches:
            batch_sz = random.randint(*self.batch_size_range)
            batch_index = tuple(random.sample(xrange(n_features), batch_sz))
            batch_indices.append(batch_index)
        #print batch_indices
        ## 2. train kmeans in parallel
        ## 2.1 set up shared memory object for each processes
        X_dir, X_path = self._persist_data(X, 'X')
        ## 2.2 split kmeans fitting in chunks and do them in parallel
        dv = Client()[:]
        dv.scatter('batch_subset', batch_indices)
        dv['subset_kmeans'] = None
        dv.apply(TriKmeansFeatures._train_models, X_path, self.n_clusters)
        feat_to_kmeans_result = dv.gather('subset_kmeans')
        self.feat_to_kmeans_ = dict(feat_to_kmeans_result.get())
        #print len(self.feat_to_kmeans)
        #print self.feat_to_kmeans_.keys()
        ## 2.3 clean up temparory files
        shutil.rmtree(X_dir)
        
    @staticmethod
    def _train_models(X_path, n_clusters):
        from sklearn.externals import joblib
        from sklearn.cluster import KMeans
        X = joblib.load(X_path)
        global subset_kmeans
        subset_kmeans = [(batch,
                          KMeans(n_clusters, random_state=0).fit(X[:,batch]))
                            for batch in batch_subset]
    def transform(self, X):
        """
        Transforming steps in serial
        """
        tri_feats = []
        for (feat_index, kmeans) in self.feat_to_kmeans_.items():
            dist_to_clusters = kmeans.transform(X[:, feat_index])
            meandist_per_cluster = np.mean(dist_to_clusters, axis=0)
            tri_feat = np.apply_along_axis(lambda row: np.maximum(0, 
                                                                  meandist_per_cluster-row),
                                                1, dist_to_clusters)
            tri_feats.append(tri_feat)
        tri_feats_X = sparse.coo_matrix(np.hstack(tri_feats))
        return tri_feats_X
    def _persist_data(self, X, Xname):
        tmstamp = time.ctime().replace(' ', '_')
        X_dir = path.abspath(path.join(self.cache_dir, tmstamp))
        os.mkdir(X_dir)
        X_path = X_dir + '/%s.pkl' % (Xname)
        joblib.dump(X, X_path)
        return (X_dir, X_path)
    def fit_transform(self, X=None):
        pass
```
