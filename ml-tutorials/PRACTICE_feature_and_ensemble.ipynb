# Feature Engineering

## Tri-Kmeans Features

```python
## tri-kmeans feature selection
class TriKmeansFeatures(BaseEstimator):
    def __init__(self, n_clusters, n_batches, batch_size_range, cache_dir, random_state=0):
        """
        n_clusters: number of clusters used in KMeans
        n_batches: number of batches extracted from original feats
        batch_size_range: size range of batch including how many original feats [low,high]
        cache_dir: the cache directory for shared memory object
        So the total extract features would be n_batches * n_clusters.
        And the batches will be extracted randomly with replacement from origna
        feats
        Clustering methods are sensitive to feature scales. Its better to normalize 
        the train and test matrix X before feeding it to clustering methods, such as
        KMeans and MiniBatchKMeans
        """
        self.n_clusters = n_clusters
        self.n_batches = n_batches
        self.batch_size_range = batch_size_range
        self.cache_dir = cache_dir
        #self.rng = check_random_state(random_state)
        #np.random.set_state(self.rng)
        random.seed(random_state)
    def fit(self, X, y=None):
        """FITTING STEPS:
        1. extract batches from original features (without redundancy)
        2. for each batch, train a clustering model, in parallel
        """
        ## 1. extract random features
        ## chance is thr that same batch of features are selected, 
        ## but it is small when n_features is large
        n_samples, n_features = X.shape
        batch_indices = []
        while len(batch_indices) != self.n_batches:
            batch_sz = random.randint(*self.batch_size_range)
            batch_index = tuple(random.sample(xrange(n_features), batch_sz))
            batch_indices.append(batch_index)
        #print batch_indices
        ## 2. train kmeans in parallel
        ## 2.1 set up shared memory object for each processes
        X_dir, X_path = self._persist_data(X, 'X')
        ## 2.2 split kmeans fitting in chunks and do them in parallel
        dv = Client()[:]
        dv.scatter('batch_subset', batch_indices)
        dv['subset_kmeans'] = None
        dv.apply(TriKmeansFeatures._train_models, X_path, self.n_clusters)
        feat_to_kmeans_result = dv.gather('subset_kmeans')
        self.feat_to_kmeans_ = dict(feat_to_kmeans_result.get())
        #print len(self.feat_to_kmeans)
        #print self.feat_to_kmeans_.keys()
        ## 2.3 clean up temparory files
        shutil.rmtree(X_dir)
        return self
        
    @staticmethod
    def _train_models(X_path, n_clusters):
        from sklearn.externals import joblib
        from sklearn.cluster import KMeans
        X = joblib.load(X_path)
        global subset_kmeans
        subset_kmeans = [(batch,
                          KMeans(n_clusters, random_state=0).fit(X[:,batch]))
                            for batch in batch_subset]
    def transform(self, X):
        """
        Transforming steps in serial
        """
        tri_feats = []
        for (feat_index, kmeans) in self.feat_to_kmeans_.items():
            dist_to_clusters = kmeans.transform(X[:, feat_index])
            meandist_per_cluster = np.mean(dist_to_clusters, axis=0)
            tri_feat = np.apply_along_axis(lambda row: np.maximum(0, 
                                                                  meandist_per_cluster-row),
                                                1, dist_to_clusters)
            tri_feats.append(tri_feat)
        tri_feats_X = sparse.coo_matrix(np.hstack(tri_feats))
        return tri_feats_X
    def _persist_data(self, X, Xname):
        tmstamp = time.ctime().replace(' ', '_')
        X_dir = path.abspath(path.join(self.cache_dir, tmstamp))
        os.mkdir(X_dir)
        X_path = X_dir + '/%s.pkl' % (Xname)
        joblib.dump(X, X_path)
        return (X_dir, X_path)
    def fit_transform(self, X, y=None):
        return self.fit(X).transform(X)
```

```python
## Since it is unsupervised feature extraction, there are two usages for tri-kmeans
## feature extraction:
## 1. build "tri-kmeans and classifier" pipeline using train data, and test the built
## model on test data
## 2. build "tri-kmeans" with both train and test data, and transform the whole data
## set with tri-kmeans. then train classifier on train data and test the model on test
## data
```

```python
## TEST (I) new features with SGDClassifier
## use tri-kmeans and sgd as a pipeline, dont touch test data during training
## of either tri-kmeans or sgd
## It seems that ipython parallel doesnt work with joblib.Parallel
tri_kmeans = TriKmeansFeatures(n_clusters = 9, n_batches=2000, 
                                batch_size_range = (2, 5), cache_dir='/tmp')
sgd = linear_model.SGDClassifier(loss='hinge', penalty='elasticnet', 
                                l1_ratio=0.9, random_state=0)
pipe = Pipeline([('tri_kmeans', tri_kmeans), 
                 ('sgd', sgd)])
alphas = np.logspace(-6, 5, 12)
cv = KFold(X_scaled.shape[0], n_folds=3, random_state=0)
scores = [cross_val_score(pipe.set_params(sgd__alpha=alpha), 
            X_scaled, y, cv = cv, n_jobs=1) 
            for alpha in alphas]
cv_scores = map(np.mean, scores)
```

## Decision Tree based Feature Extraction

```python
## decision tree based feature selection
class DecisionTreeFeatures(BaseEstimator):
    def __init__(self, n_batches, batch_size_range, cache_dir, random_state=0):
        """
        n_batches: number of batches extracted from original feats
        batch_size_range: size range of batch including how many original feats [low,high]
        cache_dir: the cache directory for shared memory object
        So the total extract features would be n_batches * n_clusters.
        And the batches will be extracted randomly with replacement from origna
        feats
        Decision Tree methods are NOT sensitive to feature scales. 
        """
        self.n_batches = n_batches
        self.batch_size_range = batch_size_range
        self.cache_dir = cache_dir
        #self.rng = check_random_state(random_state)
        #np.random.set_state(self.rng)
        random.seed(random_state)
    def fit(self, X, y):
        """FITTING STEPS:
        1. extract batches from original features (without redundancy)
        2. for each batch, train a clustering model, in parallel
        """
        ## 1. extract random features
        ## chance is thr that same batch of features are selected, 
        ## but it is small when n_features is large
        n_samples, n_features = X.shape
        batch_indices = []
        while len(batch_indices) != self.n_batches:
            batch_sz = random.randint(*self.batch_size_range)
            batch_index = tuple(random.sample(xrange(n_features), batch_sz))
            batch_indices.append(batch_index)
        #print batch_indices
        ## 2. train decision tree in parallel
        ## 2.1 set up shared memory object for each processes
        X_dir, X_path = self._persist_data(X, 'X')
        y_dir, y_path = self._persist_data(y, 'y')
        ## 2.2 split kmeans fitting in chunks and do them in parallel
        dv = Client()[:]
        dv.scatter('batch_subset', batch_indices)
        dv['subset_dt'] = None
        dv.apply(DecisionTreeFeatures._train_models, X_path, y_path)
        feat_to_dt_result = dv.gather('subset_dt')
        #print feat_to_dt_result.get()[0]
        self.feat_to_dt_ = dict(feat_to_dt_result.get())
        #print len(self.feat_to_kmeans)
        #print self.feat_to_kmeans_.keys()
        ## 2.3 clean up temparory files
        shutil.rmtree(X_dir)
        return self
        
    @staticmethod
    def _train_models(X_path, y_path):
        from sklearn.externals import joblib
        from sklearn.tree import DecisionTreeClassifier
        X = joblib.load(X_path)
        y = joblib.load(y_path)
        global subset_dt
        subset_dt = [(batch,
                          DecisionTreeClassifier(max_depth=5).fit(X[:,batch], y))
                            for batch in batch_subset]
    def transform(self, X):
        """
        Transforming steps in serial
        """
        dt_feats = []
        for (feat_index, dt) in self.feat_to_dt_.items():
            dt_feat = dt.predict_proba(X[:,feat_index])
            dt_feats.append(dt_feat)
        dt_feats_X = sparse.coo_matrix(np.hstack(dt_feats))
        return dt_feats_X
    def _persist_data(self, X, Xname):
        tmstamp = time.ctime().replace(' ', '_')
        X_dir = path.abspath(path.join(self.cache_dir, tmstamp))
        if not os.path.exists(X_dir):
            os.mkdir(X_dir)
        X_path = X_dir + '/%s.pkl' % (Xname)
        joblib.dump(X, X_path)
        return (X_dir, X_path)
    def fit_transform(self, X, y):
        return self.fit(X, y).transform(X)
```

```python
## test decision tree based feature extraction
dt_feat_extractor = DecisionTreeFeatures(n_batches=2000, 
                                batch_size_range = (1, 2), cache_dir='/tmp')
#dt_feat_extractor.fit(X, y)
feats = dt_feat_extractor.fit_transform(X, y)
print feats.shape
```

```python
## test SGD with new extracted features
alphas = np.logspace(-6, 5, 12)
scores = []
for alpha in alphas:
    cv_scores = []
    cv = KFold(X.shape[0], n_folds=3, random_state=0)
    for train, test in cv:
        train_X, train_y = X[train, :], y[train]
        test_X, test_y = X[test, :], y[test]
        dt_feater = DecisionTreeFeatures(n_batches=2000, 
                                batch_size_range = (1, 2), cache_dir='/tmp')
        train_feats = dt_feater.fit_transform(train_X, train_y)
        test_feats = dt_feater.transform(test_X)
        sgd = linear_model.SGDClassifier(loss='hinge', penalty='elasticnet', 
                                l1_ratio=0.9, random_state=0)
        sgd.set_params(alpha=alpha).fit(train_feats, train_y)
        cv_scores.append(sgd.score(test_feats, test_y))
    scores.append(np.mean(cv_scores))
print scores
```
