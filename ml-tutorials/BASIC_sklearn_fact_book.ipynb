- ***Reguarlization***

    - **Regularization** can be viewed as to *shrink* model coefficients (e.g. of linear models) to zero. 

    - Ideally this will reduce the correlation between two randomly chosen set of observations, and thus reduce the variance of model estimate and improve generalization. It is specially useful for high-dimensional data.

    - *Coefficient Regularization* is also useful as a *feature seleciton* method in high-dimensional space learning. It is useful to find linear subspace of the high-d data when used with linear models.

    - **Ridge Regularization** penalizes `l2` norm and will reduce the coefficients of less-informative features but wont be able to set them to exactly zero.

    - **Lasso Regularization** penalizes `l1` norm and will be able to find sparse solutions. Its implementation uses `coordinate decent` method, which is efficient on **large** datasets.

    - **LassoLars** in sklearn is similar to `Lasso` but it is more efficient for problems in which the weight vector estimated is VERY SPARSE (i.e., problems with very few observations)

- **Linear Model for Classification**

    - Applying regression models to classification will give too much weight on points far from the decision boundary, due to the cost function used in regression models.

    <img src="http://scikit-learn.org/stable/_images/plot_logistic_11.png">

    - Multiclassification in sklearn is usually done by fitting one-versus all classifiers and using voting for final decision

    - *parameters* for `LogisticRegression` model: `C`:  the weights of fitting error over regularization, `penalty` = `l2` gives shrinkage, whereas `l1` gives sparsity

    - *parameters* for `LinearSVC`:  a small value for `C` means the margin is calculated using many or all of the observations around the separating line (more regularization); a large value for `C` means the margin is calculated on observations close to the separating line (less regularization) - **data normalizatino is crucial to svm preformances**

**Comparison of different cross validation generator**

- **KFold (n, k)** Split it K folds, train on K-1 and then test on left-out

- **StratifiedKFold (y, k)** It preserves the class ratios / label distribution within each fold.

- **LeaveOneOut (n)** Leave one observation out

- **LeaveOneLabelOut (labels)** Takes a label array to group observations

- **Clustering for feature engineering and classification**

    - `Tri KMeans` for finding features in deep learning. see the paper here http://www.stanford.edu/~acoates/papers/coatesleeng_aistats_2011.pdf. Implementation in sklearn includes using 
        - `KMeans`, `MiniBatchKMeans`, `KMeans.fit`, `KMeans.predict` (nearst cluster center), `KMeans.transform` (distance to different centers), `np.choose`
    - `KMeans` for classification and thus feature calibration (via probability value)
        - useful implementations: `metrics.confusion_matrix`, `metrics.adjusted_mutual_info_score`, `KMeans.predict`
    - *vector quantization*: Clustering in general and KMeans, in particular, can be seen as a way of choosing a small number of exemplars to compress the information.
    - *connectivity-constrained clustering* - `Ward` or `WardAgglomerative` (bottom-up method that is usually more efficient than topdown like kmeans) - useful for feature allomeration in image data, use `grid_to_graph` in `sklearn.feature_extraction.image`

**PCA vs ICA vs LDA(Linear Discriminant Analysis)**

- *PCA*
    - Focus on uncorrelated and Gaussian components
    - Second-order statistics
    - Orthogonal transformation
 

- *ICA*
    - Focus on independent and non-Gaussian components
    - Higher-order statistics
    - Non-orthogonal transformation
 
