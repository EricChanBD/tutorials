![overall cost](http://ufldl.stanford.edu/wiki/images/math/4/5/3/4539f5f00edca977011089b902670513.png)

- Note: Usually weight decay is not applied to the bias terms , as reflected in our definition for J(W,b). Applying weight decay to the bias units usually makes only a small difference to the final network, however.
- This cost function above is often used both for *classification* and for *regression* problems. For classification, we let y = 0 or 1 represent the two class labels (recall that the sigmoid activation function outputs values in [0,1]; if we were using a tanh activation function, we would instead use -1 and +1 to denote the labels). For regression problems, we first scale our outputs to ensure that they lie in the [0,1] range (or if we were using a tanh activation function, then the [ − 1,1] range).

#### 2. initialization
- Our goal is to minimize J(W,b) as a function of W and b. To train our neural network, we will initialize each parameter  and each  to a small random value near zero (say according to a Normal(0,ε2) distribution for some small ε, say 0.01), and then apply an optimization algorithm such as batch gradient descent.
- Since J(W,b) is a non-convex function, gradient descent is susceptible to local optima; however, in practice gradient descent usually works fairly well. 
- Finally, note that it is important to initialize the parameters randomly, rather than to all 0's. If all the parameters start off at identical values, then all the hidden layer units will end up learning the same function of the input (more formally,  will be the same for all values of i, so that  for any input x). The random initialization serves the purpose of *symmetry breaking*.
#### 3. backpropagation and learning
- backpropagation algorithm is a way to compute these partial derivatives.
- doing optimization by using a library: The main thing you need to provide to these advanced optimization algorithms is that for any , you have to be able to compute  and . These optimization algorithms will then do their own internal tuning of the learning rate/step-size  (and compute its own approximation to the Hessian, etc.) to automatically search for a value of  that minimizes, such as L-BFGS and conjugate gradient.

```python
def shared_data_xy(data):
    """
    Function that loads data into shared variable.
    The reason we store our dataset in shared variables 
    is to allow theano to copy it into GPU.
    Since copying data into the GPU is slow, copying a minibatch 
    everytime is needed (the default behavior is the data is not in a shared variable) 
    would lead to a large decrease in performance.
    """
    data_x, data_y = data
    shared_x = theano.shared(np.asarray(data_x, dtype=theano.config.floatX))
    shared_y = theano.shared(np.asarray(data_y, dtype=theano.config.floatX))
    
    return shared_x, T.cast(shared_y, 'int32')
```

```python
train_set_x, train_set_y = shared_data_xy(train_small)
valid_set_x, valid_set_y = shared_data_xy(valid_small)
test_set_x, test_set_y = shared_data_xy(test_small)
```

### Logistic Regression MSGD

```python
## the class only constructs graph model in theano - those symbolic variables
class LogisticRegression(object):
    def __init__(self, nin, nout, X = None, y = None, W = None, b = None):
        ## X, W, b are passed as default parameters, as they could 
        ## be shared by other models/layers in a deep network
        ## model inputs
        ## y is the target vector (must be from 0 to class_label - 1)
        ## but most of time, target y will be dynamically bound in minibatch
        ## so it is a free variable
        self.X = X or T.matrix('X')
        self.y = y or T.ivector('y')
        ## model parameters
        self.W = W or theano.shared(value = np.random.normal(scale = 0.3, size = (nin, nout)))
        self.b = b or theano.shared(value = np.zeros((nout, )))
        self.params = (self.W, self.b)
        ## model outputs
        ## use X.W instead of W.X to ensure that b is boardcasted correctly
        self.p_y_given_x = T.nnet.softmax(T.dot(self.X, self.W) + self.b)
        self.y_pred = T.argmax(self.p_y_given_x, axis = 1)
        ## model cost, error and score
        self.cost = -T.mean(T.log(self.p_y_given_x)[T.arange(self.y.shape[0]), self.y]) # negative log conditional likelihood
        self.error = T.mean(T.neq(self.y_pred, self.y)) # zero-one-loss
```

```python
## construct functions based on symbols in certain formular
## those are the only things used in certain optimization algorithm such as 
## minibatch stochastic gradient decent (MSGD) with early stopping

## fascade 1 of formula - get train error and update parameters
def build_train_model(formula, data, batch_size, learning_rate = None):
    ## data = (train_X, train_y)
    learning_rate = learning_rate or 0.01
    index = T.lscalar('index')
    gparams = T.grad(formula.cost, formula.params)
    updates = [(param, param - learning_rate * gparam) 
                    for (param, gparam) in zip(formula.params, gparams)]
    data_X, data_y = data
    givens = {  formula.X : data_X[index*batch_size:(index+1)*batch_size], 
                formula.y : data_y[index*batch_size:(index+1)*batch_size]}
    train_model = theano.function(inputs = [index], 
                    outputs = formula.error, 
                    updates = updates, 
                    givens = givens)
    return train_model

## fascade 2 of formula - get validation/test error
def build_evaluate_model(formula, data):
    ## data = (valid_X, valid_y) or (test_X, test_y)
    data_X, data_y = data
    givens = {  formula.X : data_X,
                formula.y : data_y}
    evaluate_model = theano.function(inputs = [], 
                    outputs = formula.error, givens = givens)
    return evaluate_model
```

```python
## construct the instance and give it a go - SIMPLE TEST
batch_size = 600
lr = LogisticRegression(nin = 28 * 28, nout = 2)
train_model = build_train_model(lr, (train_set_x, train_set_y), 600)
valid_model = build_evaluate_model(lr, (valid_set_x, valid_set_y))
test_model = build_evaluate_model(lr, (test_set_x, test_set_y))
print train_model(0)
print valid_model()
print test_model()
```

```python
import copy

def msgd_optimize(nepochs, batch_size, formula, train_data, valid_data):
    """
    nepochs : number of epochs for the minibatch update 
    batch_size : the size of the mini-batch
    formula : the learning model symbolic variable graph
    train_data, valid_data : training, validation data
    """
    ## hyper-parameter setting
    nsamples, nfeats = train_data[0].get_value(borrow = True).shape
    nclasses = len(np.unique(train_data[1].eval()))
    nbatches = nsamples / batch_size
    learning_rate = 0.01
    patience = 3000
    patience_increase = 2 ## geomatrically increase
    improvement_threshold = 0.995
    validate_frequency = min(nbatches, patience / 2)
    
    ## build drive functions
    train_model = build_train_model(formula, train_data, batch_size, learning_rate)
    validate_model = build_evaluate_model(formula, valid_data)
    
    ## house keeping 
    best_params = None
    best_validation_error = np.inf
    
    ## iterative update - epochs -> batches
    epoch = 0
    done_looping = False
    while epoch < nepochs and not done_looping:
        #print '---- epoch = %i ----' % epoch
        for ibatch in xrange(nbatches):
            #print '---- epoch = %i batch = %i ----' % (epoch, ibatch)
            ## get the iteration number : iteration = epoch * batch_size + batch_index
            iteration = nbatches * epoch + ibatch
            ## update model parameter
            train_error = train_model(ibatch)
            ## check validation error if possible
            if iteration % validate_frequency == 0:
                validation_error = validate_model()
                if validation_error < best_validation_error:
                    if validation_error <= best_validation_error * improvement_threshold:
                        patience = max(patience, iteration * patience_increase)
                    best_validation_error = validation_error
                    best_params = copy.deepcopy([param.get_value(borrow=True) for param in formula.params])
                print 'epoch %i, train_error = %f, validation_error = %f' % (epoch, train_error, validation_error)
            ## check for early stopping
            if iteration > patience:
                done_looping = True
                print 'EARLY STOPPING'
                break
        epoch += 1
    print 'best validation error: %f' % best_validation_error
    print 'best param found: ', best_params 
```

```python
lr = LogisticRegression(nin = 28 * 28, nout = 2)
msgd_optimize(nepochs = 20, batch_size = 600, formula = lr, 
            train_data = (train_set_x, train_set_y),
            valid_data = (valid_set_x, valid_set_y))
test_model = build_evaluate_model(lr, (test_set_x, test_set_y))
print 'test error: ', test_model()
```

### Logistic Regression Conjugate Gradient (CG)

```python
## logistic regressoin model
class ParameterizedLogisticRegression(object):
    def __init__(self, nin, nout, X = None, y = None, theta = None):
        ## data information
        self.nin = nin
        self.nout = nout
        self.X = X or T.matrix('X')
        self.y = y or T.ivector('y')
        ## parameters
        self.theta = theta or theano.shared(value = np.random.normal(scale = 0.03, size = (nin*nout+nout,)),
                        name = 'theta', 
                        borrow = True)
        self.W = self.theta[:nin*nout].reshape((nin, nout))
        self.b = self.theta[nin*nout:]
        self.params = [self.theta]
        ## predictions
        self.p_y_given_x = T.nnet.softmax(T.dot(self.X, self.W) + self.b)
        self.y_pred = T.argmax(self.p_y_given_x, axis = 1)
        ## error and cost
        self.cost = -T.mean(T.log(self.p_y_given_x)[T.arange(self.y.shape[0]), self.y]) ## NLL
        self.error = T.mean(T.neq(self.y_pred, self.y)) ## zero-one loss
        ## gradient
        self.gparams = T.grad(self.cost, self.theta)
```

```python
"""
Interfaces of the functions used by the cg optimizer:
def train_fn(theta_value):
    ## return error
    pass
def train_grad(theta_value):
    ## return gradient of error wrt theta
    pass
def callback(theta_value):
    ## check validation error
    pass

Use FACTORY PATTERN to build those functions, because those functions
need both formula and data information, whereas they are supposed to be
paramter-free (except theta_value) in the optimizers
"""

def build_train_fn_grad(formula, train_data):
    """
    return train_fn and train_grad functions
    """
    train_X, train_y = train_data
    _fn = theano.function(inputs = [], 
                    outputs = formula.cost,
                    givens = {formula.X : train_X,
                              formula.y : train_y})
    _grad = theano.function(inputs = [], 
                    outputs = formula.gparams,
                    givens = {formula.X : train_X,
                              formula.y : train_y})
    def _train_fn(theta_value):
        formula.theta.set_value(theta_value, borrow = True)
        return _fn()
    def _train_grad(theta_value):
        formula.theta.set_value(theta_value, borrow = True)
        return _grad()
    return (_train_fn, _train_grad)

def build_callback(formula, train_data, validate_data):
    """
    return callback function which prints train and valiation error
    """
    train_X, train_y = train_data
    validate_X, validate_y = validate_data
    _train_error = theano.function(inputs = [], 
                    outputs = formula.error,
                    givens = {formula.X : train_X,
                              formula.y : train_y})
    _validate_error = theano.function(inputs = [],
                    outputs = formula.error,
                    givens = {formula.X : validate_X,
                              formula.y : validate_y})
    def _callback(theta_value):
        formula.theta.set_value(theta_value, borrow = True)
        train_error = _train_error()
        validate_error = _validate_error()
        print 'Train Error %f, Validate Error %f' % (train_error, validate_error)
    return _callback
```

```python
lr = ParameterizedLogisticRegression(nin = 28 * 28, nout = 2)
train_fn, train_grad = build_train_fn_grad(lr, (train_set_x, train_set_y))
print train_fn(np.zeros(((28 * 28 + 1) * 2)))
print train_grad(np.zeros(((28 * 28 + 1) * 2)))
callback = build_callback(lr, (train_set_x, train_set_y), (valid_set_x, valid_set_y))
callback(np.zeros(((28 * 28 + 1) * 2)))
```
