```python
## get 10000 patches of size 8 x 8 randomly
def random_patches(n_patches, images, patch_size):
    n_images, n_rows, n_cols = images.shape
    patch_row, patch_col = patch_size
    patches = np.zeros((n_patches, patch_row, patch_col))
    image_indices = rng.randint(0, n_images, size = n_patches)
    row_indices = rng.randint(0, n_rows - patch_row, size = n_patches)
    col_indices = rng.randint(0, n_cols - patch_col, size = n_patches)
    for i in xrange(n_patches):
        im, ir, ic = image_indices[i], row_indices[i], col_indices[i]
        patches[i] = images[im, ir:ir+patch_row, ic:ic+patch_col]
    return patches

patches = random_patches(10000, images, (8, 8))
feats = patches.reshape((10000, 64))
```

```python
## Normalize the data - squash data to [0.1, 0.9] since we use sigmoid as the activation function
## STEPS: (image-wise)
## 1. remove DC (mean of all imageS) - for each image separately
image_means = np.mean(feats, axis = 1)
print image_means.shape
feats = (feats.T - image_means).T
print feats.shape, np.mean(feats)
## 2. Truncate to +/- 3 standard deviations and scale to -1 and 1 - for each image separately
overall_std = 3 * np.std(feats)
feats = np.maximum(np.minimum(feats, overall_std), -overall_std) / overall_std
print np.min(feats), np.max(feats)
## 3. Rescale from [-1, 1] to [0.1, 0.9] - for each image separately
feats = (feats + 1) * .4 + .1
print np.min(feats), np.max(feats), feats.shape
```

## Sparse Auto Encoder

```python
import theano
import theano.tensor as T
```

```python
class SparseAutoEncoder(object):
    def __init__(self, n_visible, n_hidden, rao = 0.5, 
                     wd_coeff= 0.5, s_coeff = 0.2,
                    X = None, theta = None):
        """
        Dimension of parameters: 
            W (n_visible x n_hidden), bvis (n_visible), bhid (n_hidden)
            theta = [W.flatten + bvis.flatten + bhid.flatten]
        """
        ## meta parameters
        self.n_visible = n_visible
        self.n_hidden = n_hidden
        self.rao = rao ## sparsity threshold
        self.wd_coeff = wd_coeff ## weight decay term penalty coefficient
        self.s_coeff = s_coeff ## sparsity term penalty coefficient
        self.np_rng = np.random.RandomState(0)
        ## inputs
        self.X = X or T.matrix('X')
        ## parameters
        if not theta:
            theta_value = self.np_rng.uniform(
                            low = -np.sqrt(6. / (n_visible + n_hidden)),
                            high = np.sqrt(6. / (n_visible + n_hidden)),
                            size = ((n_visible + 1) * (n_hidden + 1)-1, ))
            self.theta = theano.shared(value = np.asarray(theta_value, dtype = theano.config.floatX), 
                    name = 'theta', 
                    borrow = True)
        else:
            self.theta = theta
        self.W = self.theta[:n_visible*n_hidden].reshape((n_visible, n_hidden))
        self.W_prime = self.W.T # tied weight
        self.bvis = self.theta[n_visible*n_hidden:n_visible*(n_hidden+1)]
        self.bhid = self.theta[n_visible*(n_hidden+1):]
        ## predictions
        self.Y = T.nnet.softmax(T.dot(self.X, self.W) + self.bhid)
        self.Z = T.nnet.softmax(T.dot(self.Y, self.W_prime) + self.bvis)
        ## cost - 3 terms in the cost 
        ## 1. mean squared error 
        ## 2. weigth decay - only on W terms (excluding b terms)
        ## 3. sparsity
        self.squared_error = T.mean((self.Z - self.X) ** 2)
        self.weight_decay = T.mean(self.W ** 2) 
        sparsity = T.mean(self.Y, axis = 0) ## it is a vector of n_hidden
        self.sparsity_penalty = T.mean(self.rao*(np.log(self.rao)-T.log(sparsity)) + 
                                      (1-self.rao)*(np.log(1-self.rao)-T.log(1.-sparsity)))
        self.cost = self.squared_error + self.weight_decay * self.wd_coeff + self.sparsity_penalty * self.s_coeff
        ## gradient
        self.gparams = T.grad(self.cost, wrt = self.theta)
```

```python
def share_data(data):
    shared_data = theano.shared(np.asarray(data, dtype = theano.config.floatX), borrow = True)
    return shared_data

## build factory method for optimizer by binding formula with data
def build_f_fgrad(formula, data):
    data = share_data(data)
    cost_value = theano.function(inputs = [], 
                    outputs = formula.cost, 
                    givens = {formula.X : data})
    cost_prime = theano.function(inputs = [],
                    outputs = formula.gparams, 
                    givens = {formula.X : data})
    def _f(theta_value):
        formula.theta.set_value(theta_value, borrow = True)
        return cost_value()
    def _fprime(theta_value):
        formula.theta.set_value(theta_value, borrow = True)
        return cost_prime()
    return (_f, _fprime)
def build_callback(formula, data):
    data = share_data(data)
    _train_error = theano.function(inputs = [],
                    outputs = formula.cost,
                    givens = {formula.X : data})
    def _callback(theta_value):
        formula.theta.set_value(theta_value, borrow = True)
        print 'train error:',  _train_error()
    return _callback

#formula = SparseAutoEncoder(n_visible = 8 * 8, n_hidden = 50)
#f, fprime = build_f_fgrad(formula, feats)
#print f(np.zeros(65 * 51-1))
#print fprime(np.zeros(65 * 51-1))
```

```python
## use scipy.optimize.cf_min to optimize the sparse auto encoder
from scipy import optimize

n_visible = feats.shape[1]
n_hidden = 25
n_iters = 100
theta0 = np.concatenate([np.random.uniform( low = -np.sqrt(6. / (n_visible + n_hidden)),
                            high = np.sqrt(6. / (n_visible + n_hidden)),
                            size = (n_visible * n_hidden, )),
            np.zeros(n_visible), 
            np.zeros(n_hidden)])
formula = SparseAutoEncoder(n_visible = n_visible, n_hidden = n_hidden)
f, fprime = build_f_fgrad(formula, feats)
callback = build_callback(formula, feats)

best_theta = optimize.fmin_cg(
    f=f, 
    x0=theta0, 
    fprime=fprime, 
    maxiter=n_iters, 
    callback=callback)
```

```python
## display W matrix
formula.theta.set_value(best_theta, borrow = True)
W_value = formula.W.eval().T
print W_value.shape

ncols = 4
nrows = W_value.shape[0]/ncols
fig, axes = subplots(ncols = ncols, nrows = nrows, figsize = (1 * ncols, 1 * nrows))
axes = axes.flatten()

for i, ax in enumerate(axes):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.imshow(W_value[i].reshape(8, 8))
```
